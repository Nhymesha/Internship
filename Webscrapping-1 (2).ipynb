```python
#send a get request to the server to get a source code of the page
from bs4 import BeautifulSoup
import requests
IMDB=requests.get('https://www.imdb.com/chart/top/')
```

```python
IMDB
```

```python
#extract all the text that you received from the GET request  
#convert the text to a beautiful soup object
  

IMDB_soup=BeautifulSoup(IMDB.content)
print(IMDB_soup)
```

```python
scraped_movies=IMDB_soup.find_all('td', class_="titleColumn")   #used to extract the data related to movies
scraped_movies
movies=[]    # create Empty list to store the output
for movie in scraped_movies:    #For loop that iterates over all the <td> tags
    movie=movie.get_text().replace('\n', " ")  #using get_text to extract the text and replace to replace the data
    movie=movie.strip(" ")
    
    movies.append(movie)   #appending the movie to the output list
    
print(movies)
```

```python
scraped_ratings=IMDB_soup.find_all('td', class_="ratingColumn imdbRating") #used to extract the text related to ratings
scraped_ratings
ratings=[]     #create Empty list to store the output
for rating in scraped_ratings:  ##For loop that iterates over all the <td> tags
    rating=rating.get_text().replace('\n', " ")   # #using get_text to extract the text and replace to replace the data
    ratings.append(rating)  #appending the movie to the output list
print(ratings)
```

```python
#repeat the same method which was done as earlier

Scarp_year=IMDB_soup.find_all('span', class_="secondaryInfo") 
Scarp_year
years=[]    
for year in Scarp_year:  
    year=year.get_text().replace("/span", " ")
    year=year.strip("()")
    years.append(year)
print(years)
```

```python
import pandas as pd
data=pd.DataFrame()   #creating a dataframe
```

```python
#Accessng the columns to dataframe
data["Movie Names"]= movies
data['Ratings']= ratings
data["Year of Release"]=years
data
```

```python
#print top 100 movies from the list
data.head(100)
```
