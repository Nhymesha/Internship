Write a python program to scrape mentioned details from dineout.co.in 
Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
#	We will first install and import our required libraries.
!pip install bs4
!pip install requests
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
#perform get request to the url

  
# extract all the text that you received from the GET request  

request=requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
request
```

```python
#convert the text to a beautiful soup object to download the page content
dine_soup=BeautifulSoup(request.content)
dine_soup
```

```python
#creating empty list to save the result of titles
titles=[]
for restnt in dine_soup.find_all('a', class_="restnt-name ellipsis"): #For loop that iterates over all the tags of "a"
    titles.append(restnt.text)  #appending the restnt to the output list
    
titles
```

```python
location=[]

for loc in dine_soup.find_all("div", class_="restnt-loc ellipsis"):   #For loop that iterates over all the tags
    location.append(loc.text)  #appending the loc to the output list
    
location
```

```python
#repeating the same action as earlier

cuisine_rate=[]

for i in dine_soup.find_all("div", class_="detail-info"): 
    i=i.get_text().replace('for 2 (approx)', " ").replace("  ", " ")
    cuisine_rate.append(i)
    
cuisine_rate
    
```

```python
#extracting the urls using following code

URL=[]

for url in dine_soup.find_all("a", class_="restnt-name ellipsis"):
    URL.append(url["href"])
    
URL
   
```

```python
import pandas as pd
```

```python
#make a data frame my merging all the above columns
data=pd.DataFrame({"Restaurant name":titles, "Location":location, "Rate & Cuisine":cuisine_rate, "URL":URL})
data
```

Write a python program to scrape the details of top publications from Google Scholar from
https://scholar.google.com/citations?view_op=top_venues&hl=en
i) Rank
ii) Publication
iii) h5-index
iv) h5-median

```python
#sending a get request to the server to get the source code
request1=requests.get("https://scholar.google.com/citations?view_op=top_venues&hl=en")
request1
```

```python
#Below code downloads the page content
Soup=BeautifulSoup(request1.content)
Soup
```

```python
#creating list to store output

Rank=[]
for items in Soup.find_all("td", class_="gsc_mvt_p"): #using for loop to iterate the td tags
    items=items.get_text().replace(".", " ")
    Rank.append(items) #append the data
    
Rank   #calling the ouput
```

```python
Publication=[]

for i in Soup.find_all("td", class_="gsc_mvt_t"):
    Publication.append(i.text)
    
    
    
Publication
```

```python
H5_index=[]
for index in Soup.find_all("a", class_="gs_ibl gsc_mp_anchor"):
    H5_index.append(index.text)
    
H5_index
```

```python
#extracing the median using below syntax
H5_median=[]
for median in Soup.find_all("span", class_="gs_ibl gsc_mp_anchor"):
    H5_median.append(median.text)
    
H5_median
```

```python
import pandas as pd
```

```python
#creating a list of all the data

print(f'''
Rank: {Rank}
Publication: {Publication}
h5_index: {H5_index}
h5_median: {H5_median}
''')
```

OR

```python
Google_scholar=pd.DataFrame({"Rank": Rank, "Publication": Publication, "h5_index": H5_index, "h5_median": H5_median})
Google_scholar
```

Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world :
i) Headline
ii) Time
iii) News Link

```python
request3=requests.get(" https://www.cnbc.com/world/?region=world")
request3
```

```python
Downloading the webpage content
News_details=BeautifulSoup(request3.content)
News_details
```

```python
#Extrcting news headlines 
Headline=[]
for news in News_details.find_all("a", class_="LatestNews-headline"):
    Headline.append(news.text)
    
Headline
```

```python
#extracting the time from content
Time=[]
for time in News_details.find_all("time", class_="LatestNews-timestamp"):
    Time.append(time.text)
    
Time
```

```python
#extracting the urls of the page

urls=[]

for h in News_details.find_all("a", class_="LatestNews-headline"):
    urls.append(h["href"])

urls    
```

```python
import pandas as pd
```

```python
News=pd.DataFrame({"Headline":Headline, "Time":Time, "urls":urls })
News
```
